```python
import cv2
import numpy as np

def find_color_data(card):
    hsv = cv2.cvtColor(card, cv2.COLOR_BGR2HSV)
    h, w, _ = card.shape
    
    center_region = hsv[h//10:9*h//10, w//10:9*w//10]
    
    avg_color = np.mean(center_region, axis=(0,1))
    
    hue = avg_color[0]
    sat = avg_color[1]
    val = avg_color[2]
    
    return {"h": hue, "s": sat, "v": val}
```

```python
card_colors = ["red", "blue", "green", "yellow", 
    "black", "white", "orange", "pink", "joker", "quest"]

for color in card_colors:
    filename = f"../data/card_{color}.png"
    img = cv2.imread(filename)
    if img is None: continue

    data = find_color_data(img)
    
    # Print formatted output
    print(f"{color:<10} | H: {data['h']:<6.1f} S: {data['s']:<6.1f} V: {data['v']:<6.1f}")
```
